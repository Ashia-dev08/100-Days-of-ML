```python
df_movies=pd.read_excel("Untitled Folder 1/movies_db.xlsx","movies")
df_movies.head(4)
```

```python
df_financials=pd.read_excel("Untitled Folder 1/movies_db.xlsx",'financials')
df_financials.head(4)
```

```python
#excel converter

def standardize_currency(curr):
    if curr=="$$" or curr == "Dollars":
        return"USD"
    return curr

df_financials=pd.read_excel("Untitled Folder 1/movies_db.xlsx",'financials',converters={'currency':standardize_currency })
df_financials.head(4)
```

```python
#MERGING BOTH THE SHEETS
df_merged=pd.merge(df_movies,df_financials,on="movie_id")
df_merged.head()
```

```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="merged",index= False)
```
